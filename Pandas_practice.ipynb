```python
df2 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Math_Score': [85, 92, 78, 88, 95],
    'Science_Score': [90, 85, 88, 92, 87],
    'Grade_Level': [9, 10, 11, 12, 10]
})
df2.groupby('Grade_Level')['Math_Score'].mean()
```

```python
df3 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve', 'Frank'],
    'Age': [15, 16, np.nan, 17, np.nan, 15],
    'Grade_Level': [9, 10, 11, 12, 9, 9]})
df3
df3['Age'].isnull().sum()
df3['Age'] = df.groupby('Grade_Level')['Age'].transform(
    lambda x: x.fillna(x.median())
)
df3
df3.drop('Name', axis=1, inplace=True)
df3
```

```python
df = pd.DataFrame({
    'Revenue': [1500, 800, 3000, 6000, 2500, 4500, 900],
    'Region': ['North', 'South', 'East', 'North', 'South', 'West', 'North'],
    'Product_Category': ['Electronics', 'Clothing', 'Food', 'Clothing','Electronics', 'Food', 'Clothing']
})
after_filter = df[(df['Revenue'].between(1000,5000))&
            (df['Region'].isin(['North', 'South'])) &
            (df['Product_Category'] != 'Electronics')]
after_filter
df
```

```python
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David'],
    'Department': ['IT', 'HR', 'IT', 'Finance'],
    'Salary': [70000, 50000, 80000, 60000]
})
after_sort = df.sort_values(by=['Department','Salary'],ascending=[True,False])
after_sort
```

```python
prod = pd.DataFrame({
    'product_id': [101, 102, 103, 104],
    'product_name': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'category': ['Electronics', 'Accessories', 'Accessories', 'Electronics']
})

sales = pd.DataFrame({
    'product_id': [101, 101, 102, 103, 105],
    'sale_date': ['2024-01-01', '2024-01-02', '2024-01-01', '2024-01-03', '2024-01-04'],
    'quantity': [2, 1, 5, 3, 2],
    'revenue': [2400, 1200, 125, 135, 60]
})
prod
sales
In_join = pd.merge(prod, sales, on='product_id',how='inner')
LF_join = pd.merge(prod, sales, on='product_id',how='left')
RG_join = pd.merge(prod, sales, on='product_id',how='right')
OT_join = pd.merge(prod, sales, on='product_id',how='outer')
In_join
LF_join
RG_join
OT_join
```
